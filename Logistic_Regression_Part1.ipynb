```r
# recode the status variable into 1's and 0's

mydata$cancer_status <- recode(mydata$status, 'M' = 1, 'B' = 0)
table(mydata$status)
table(mydata$cancer_status)
```

```r
mydata %>%
ggplot() +
  aes( x = radius, y = cancer_status) +
  geom_jitter(aes(col=status),height = 0.05, width = 0.01, alpha = 0.5)
```

```r
# prompt: can you add the logistic regression curve into my plot and col the curve in red

# Fit a logistic regression model


# Add the logistic regression curve to the plot
mydata %>%
  ggplot() +
  aes(x = radius, y = cancer_status) +
  geom_jitter(aes(col = status), height = 0.05, width = 0.01, alpha = 0.5) +
  geom_smooth(method = "glm",
              method.args = list(family = "binomial"),
              se = FALSE, color = "red")
```

```r
my_model <- glm(cancer_status ~ radius, data = mydata, family = binomial)
summary(my_model)
```

fitted curve:

$ \hat p = \frac{e^{-15.2+1.03radius}}{1+e^{-15.2+1.03radius}}$

how to interpret the 'slope'? (our 'slope' is 1.03)

- first, acknowledge the sign of the 'slope' parameter
  - in this case, the positive sign of the 'slope' indicates that probability increases as we move from small to large radius values
- In order to think about rate of change in probability, we need to make some judgements on the x values we use.

```r
my_new_data <- data.frame(radius = c(14,15,20,21))
my_predictions <- predict(my_model, newdata = my_new_data, type = 'response')
my_predictions
```
